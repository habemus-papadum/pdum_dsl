# LabelSliderWidget demo (Jupyter)

Two synced integer traits, both clamped to `[0, 100]`:

- **`label_value`** — written from Python, rendered in JS as a read-only label.
- **`slider_value`** — bidirectional: dragging the slider sends the value back to Python.

```python
from widget import LabelSliderWidget

w = LabelSliderWidget(label_value=42, slider_value=10)
w  # drag the slider
```

## Observe slider moves coming from JS

Run this, then drag the slider in the cell above — each move updates the line
below **in place** (via an `IPython.display` handle) instead of printing a new line.

```python
from IPython.display import display, HTML

# A display handle we can update in place from the observe callback.
handle = display(HTML(f"slider_value: <b>{w.slider_value}</b>"), display_id=True)

def on_slider(change):
    handle.update(HTML(f"slider_value: <b>{change['new']}</b>"))

w.observe(on_slider, names="slider_value")
```

## Drive the widget from Python

Writing the traits pushes the new values into the rendered widget above.

```python
w.label_value = 75   # updates the label
w.slider_value = 30  # moves the slider (and updates the handle above)
```

```python
w.slider_value = 250  # validator clamps to 100
print(w.slider_value)
```
